Series: Series are single dimensional one column data

```python
import pandas as pd
#series from List
data =[1,2,3,4,5,6]
series=pd.Series(data)
print("series from list \n",series) # idnex  series data

#series from dictionary
dict1={'a':1,'b':2,'c':3}
dict_series=pd.Series(dict1)
print("series from dict: \n",dict_series)
```

```python
#series with manual index
data=[1,2,3]
index=[0,1,2]
pd.Series(data,index)
```

```python
#data frame --> two dimensional -->rows vs columns
#from dictionary where values are in list
dict_={'Name':['Krish','John','Jack'],
       'Age':[35,32,45],
       'Country':['Bangalore','New York','Florida']
       }
pd.DataFrame(dict_)
```

```python
#accessing data fram
dict_data=[
             {'Name':'Krish','Age':35,'Country':'Bangalore'},
             {'Name':'John','Age':32,'Country':'New York'},
             {'Name':'Jack','Age':45,'Country':'Florida'}
           ]
df=pd.DataFrame(dict_data)
print(df)

#df['Name']
#df.loc[0]['Name'] #row index wise data
df.iloc[0][0]

#at, iat  for specific search
df.at[0,'Name']  #syntax df.at[rowIndex,columnIndex]
df.iat[0,0]
```

```python
df
#adding a new column
df['Salary']=[50000,60000,70000]
df
#removing Salary
df.drop('Salary',axis=1)
df
df.drop('Salary',axis=1,inplace=True)
df
```
