```python
import pandas as pd
df=pd.read_csv('data.csv')
df.head(5)
df.tail(10)
df.describe()
```

```python
df.dtypes  #object means string
```

```python
#working with missing values
df.isnull()
#which column having missing values
df.isnull().any()
#columnwise no of missing values
df.isnull().sum()
#replacement missing values with 0
df_filled=df.fillna(0)
df_filled
#create a new column of sales_na where missing values will be filled by its mean
df['sales_na']=df['Sales'].fillna(df['Sales'].mean())
df
#renaming column name
df=df.rename(columns={'Date':'Sales_Date'})
df.head()
#change data type
df['value_nw']=df['Value'].astype(float)
df.head()
#due to NaN unable to convert int and need to fill prior to do it
df['value_nw']=df['Value'].fillna(df['Value'].mean()).astype(int)
df.head()
#apply function on a column : suppose value_nw will be double now in separate field
df['nw_value']=df['value_nw'].apply(lambda x:x*2)
df.head()
```

```python
#data aggregating and grouping
print(df.head())
#Productwise grouping and mean of Value
grouped_prod=df.groupby('Product')['Value'].mean()
grouped_prod

#grouping by multiple column and sum
grouped_columns=df.groupby(['Product','Region'])['Value'].sum()
grouped_columns
```

```python
df.head()
#apply multiple aggregate function
multi_aggr=df.groupby('Region')['Value'].agg(['sum','mean','count'])
multi_aggr
```

```python
#Merging and joining dataframe
#sample dataframe
df1=pd.DataFrame({'key':['A','B','C'],'VALUE1':[1,2,3]})
df2=pd.DataFrame({'key':['A','B','D'],'VALUE2':[4,5,6]})
pd.merge(df1,df2,on='key',how='inner')
pd.merge(df1,df2,on='key',how='outer')
pd.merge(df1,df2,on='key',how='left')
pd.merge(df1,df2,on='key',how='right')
```
